# Checking Imbalance

```python
fraud_counts = df['Class'].value_counts()
plt.figure(figsize=(6,4))
sns.barplot(x=fraud_counts.index, y=fraud_counts.values)
plt.title('Class Distribution')
plt.show()

fraud_ratio = fraud_counts[1] / fraud_counts.sum()
print(f"Fraud Ratio: {fraud_ratio:.6f}")
```

# Time Analysis for fraud transactions

```python
df['Hour'] = ((df['Time']/3600)%24).astype(int)
```

```python
hourly_stats = df.groupby('Hour')['Class'].agg(
    total_transactions='count',
    fraud_transactions='sum'
).reset_index()
```

```python
fig, ax1 = plt.subplots(figsize=(10,6))

ax2 = ax1.twinx()

ax1.plot(hourly_stats['Hour'], hourly_stats['total_transactions'],
         label='Total Transactions', color='blue', marker='o')
ax2.plot(hourly_stats['Hour'], hourly_stats['fraud_transactions'],
         label='Fraud Transactions', color='red', marker='o')

ax1.set_xlabel('Hour of Day')
ax1.set_ylabel('Total Transactions', color='blue')
ax2.set_ylabel('Fraud Transactions', color='red')
plt.title('Total vs Fraud Transactions by Hour (Dual Axis)')
plt.grid(True)

# Combine legends from both axes
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper right')

plt.show()
```

```python
fraud_rate_by_hour = df.groupby('Hour')['Class'].mean() * 100

plt.figure(figsize=(10,6))
sns.lineplot(x=fraud_rate_by_hour.index, y=fraud_rate_by_hour.values, marker='o')
plt.title('Fraud Rate (%) by Hour of the Day')
plt.xlabel('Hour')
plt.ylabel('Fraud Rate (%)')
plt.grid(True)
plt.show()
```

Fraudulent transactions appear relatively sparse but tend to occur more frequently during off-peak hours (late night or early morning, e.g., between 1 AM and 5 AM).
This indicates that fraudulent activities may be strategically timed to avoid detection during periods of low monitoring.
